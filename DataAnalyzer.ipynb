Import libraries

```python
import matplotlib.pyplot as plt
```

Utility functions

```python
def read_rewards(filename):
  f = open(filename, 'r')
  average_rewards = [float(line.strip()) for line in f.readlines()]

  f.close()

  return average_rewards

def read_metrics(filename):
  metrics = []

  f = open(filename, 'r')

  for line in f.readlines():
    l = line.strip()
    if line != "-nan\n":
      metrics.append(float(l))
  
  f.close()

  return metrics
```

# Train small hexagon average rewards 

```python
average_rewards = read_rewards("average_reward_small_hexagon.txt")
average_rewards_noise = read_rewards("average_reward_small_hexagon_noise.txt")
```

```python
plt.plot(range(1, 101, 1), average_rewards,label="Without noise")
plt.plot(range(1, 101, 1), average_rewards_noise, label="With noise")

plt.xlabel('Epochs')
plt.ylabel('Average reward')
plt.legend()
plt.show()
```
